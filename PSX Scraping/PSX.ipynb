## Stock 

```python
# Stock
stock = 'LOTCHEM'

# Beautiful Soup PSX website setup
url = 'https://dps.psx.com.pk/company/'+ stock
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36 Edg/86.0.622.63'}
page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')

# Length of list
length = len(soup.find_all("div", class_ = "stats_value"))-1

# Company Profile
company_profile = list(soup.find_all("p")[0])[0]

# Closing Price
closing_price = list(soup.find_all("div", class_= "quote__close")[0])[0]

# Change Value
change_value = list(soup.find_all("div",class_ = "change__value")[0])[0]

# Change Percent
change_percent = list(soup.find_all("div",class_ = "change__percent")[0])[0]

# 1-Year Change
one_yr_change = list(soup.find_all("div",class_ = "stats_value")[15])[1].strip()

# YTD Change
ytd_change = list(soup.find_all("div",class_ = "stats_value")[16])[1].strip()

# Free Float (%)
free_float_perc = list(soup.find_all("div", class_ = "stats_value")[length])[0]

# Free Float
free_float = list(soup.find_all("div", class_ = "stats_value")[length-1])[0]

# Shares
total_shares = list(soup.find_all("div", class_ = "stats_value")[length-2])[0]

# Market Cap
market_cap = list(soup.find_all("div", class_ = "stats_value")[length-3])[0]

# Financials (annual)
financials_annual = pd.read_html(url)[4]

# Financials (quarterly - Last 4 quarters)
financials_quarterly = pd.read_html(url)[5]

# Ratios (Annual)
ratios_annual = pd.read_html(url)[6]

# Print
print('Company Profile\n',company_profile)
print("\nClosing Price: ", closing_price)
print("\nToday's Change in Value", change_value)
print("\nToday's Change in Value (%)",change_percent)
print('\n One Year Change: ', one_yr_change)
print('\n YTD Change: ', ytd_change)
print('\n Free Float Percentage: ', free_float_perc)
print('\n Free Float: ', free_float)
print('\n Total Shares: ', total_shares)
print('\n market_cap: ', market_cap)
print('\n Financials - Annual: \n', financials_annual)
print('\n Financials - Quarterly: \n', financials_quarterly)
print('\n Ratios - Annual: \n', ratios_annual)
```
